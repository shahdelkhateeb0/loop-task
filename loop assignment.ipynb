````python
# ِِ1. extract all the countries containing the word land.
countries_with_land = []
for country in countries:
    if 'land' in country:
        countries_with_land.append(country)
print(countries_with_land)
print( "COUNTRIES WITH 'land':", countries_with_land)
````

````python
# 2.A. What are the total number of languages in the data \ 112
languages = set()

for country in countries_data:
    languages.update(country.get("languages", []))

print("THE TOTAL NUMBER OF LANGUAGES:", len(languages))
````

````python
# 2.B. Find the ten most spoken languages from the data \ English , French , Arabic
language_count = {}

for country in countries_data:
    for language in country.get('languages', []):
        if language in language_count:
            language_count[language] += 1
        else:
            language_count[language] = 1

sorted_languages = sorted(language_count.items(), key=lambda n: n[1], reverse=True)

top_10_languages = sorted_languages[:10]

print("THE 10 MOST SPOKEN LANGUAGES ARE:")
for language, count in top_10_languages:
    print(language)
````

````python
# 2.C. Find the 10 most populated countries in the world \ China, India, US


sorted_countries = sorted(countries_data, key=lambda country: country['population'], reverse=True)

most_populated_countries = [country['name'] for country in sorted_countries[:10]]

print("THE 10 MOST POPULATED COUNTRIES IN THE WORLD:")
for country in most_populated_countries:
    print(country)
````
